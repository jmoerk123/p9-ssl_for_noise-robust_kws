```python
import os
```

```python
dir_path = "/home/jacobuni/uni-projects/google_speech_commands_v2"
dir_paths = [x[0].replace("/home/jacobuni/uni-projects/google_speech_commands_v2/", "") for x in os.walk(dir_path)]
counts = {}
```

```python
dir_paths
```

```python
for path in dir_paths:
    count = 0
    # Iterate directory
    dir = dir_path + "/" + path
    if os.path.exists(dir):
        for p in os.listdir(dir):
            # check if current path is a file
            if os.path.isfile(os.path.join(dir, p)):
                count += 1
        counts[path] = count
    else:
        print(dir)
```

```python
counts
```

```python
import matplotlib.pyplot as plt
plt.plot(counts.keys(), counts.values(), '*')
```
