```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
sound_path = "/home/jacobuni/uni-projects/google_speech_commands_v2/right/0a2b400e_nohash_0.wav"
sound = sp.io.wavfile.read(sound_path)
sr = sound[0]
sound_array = sound[1]
cmap = "viridis" #"gnuplot2" "magma" "viridis" 
```

```python
fig, ax = plt.subplots()
spec,_, _, _ = ax.specgram(sound_array, Fs=sr, cmap=plt.get_cmap(cmap))
ax.axis("off")
plt.savefig("specs/right_spec.png")
```

```python
fig, ax = plt.subplots()
ax.plot(sound_array)
ax.axis("off")
plt.savefig("specs/right_speak")
```

```python
sound_array_f = np.array(sound_array, dtype=float)
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

S = librosa.feature.melspectrogram(y=sound_array_f, sr=sr, n_mels=128, fmax=8000)
fig, ax = plt.subplots()
S_dB = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr, fmax=8000, ax=ax, cmap=plt.get_cmap(cmap))
ax.axis("off")
plt.savefig("specs/log-mel.png")
# fig.colorbar(img, ax=ax, format='%+2.0f dB')
# ax.set(title='Mel-frequency spectrogram')
```

```python
fig, ax = plt.subplots()
mfcc = librosa.feature.mfcc(y=sound_array_f, sr=sr, n_mfcc=20, fmax=8000)
img = librosa.display.specshow(mfcc, x_axis='time', y_axis='mel', sr=sr, fmax=8000, ax=ax, cmap=plt.get_cmap(cmap))
ax.axis("off")
plt.savefig("specs/mel-freq.png")
# fig.colorbar(img, ax=ax, format='%+2.0f dB')
# ax.set(title='Mel-frequency spectrogram')
```
